```python
import numpy as np
import pandas as pd
import plotly.express as px
```

# <font color=9453d9>DATASET
---

```python
dataset = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/PROJECTS/WORLD CUP/FIFA - 2022.csv", sep=',')
dataset
```

# <font color=9453d9>1. SELEES COM MAIS PARTIDAS
---

```python
games = pd.DataFrame({'Seleo': dataset['Team'],
                      'Partidas Jogadas': dataset['Games Played']})
fig = px.bar(games, x='Seleo', y='Partidas Jogadas', color_discrete_sequence=px.colors.qualitative.Bold, template='seaborn')
fig.update_layout(title = 'Selees Com Mais Partidas')
fig.show()
```

# <font color=9453d9>2. SELEES COM MAIS VITRIAS
---

```python
wins = pd.DataFrame({'Seleo': dataset['Team'],
                    'Vitrias': dataset['Win']})

fig = px.bar(wins, x='Seleo', y='Vitrias', color_discrete_sequence=px.colors.qualitative.Bold, template='seaborn')
fig.update_layout(title = 'Selees Com Mais Vitrias')
fig.show()
```

# <font color=9453d9>3. SELEES COM MAIS EMPATES
---

```python
draws = pd.DataFrame({'Seleo': dataset['Team'],
                      'Empates': dataset['Draw']})
draws.sort_values('Empates', ascending=False, inplace=True)

fig = px.bar(draws, x='Seleo', y='Empates', color_discrete_sequence=px.colors.qualitative.Bold, template='seaborn')
fig.update_layout(title = 'Selees Com Mais Empates')
fig.show()
```

# <font color=9453d9>4. SELEES COM MAIS DERROTAS
---

```python
losses = pd.DataFrame({'Seleo': dataset['Team'],
                       'Derrotas': dataset['Loss']})
losses.sort_values('Derrotas', ascending=False, inplace=True)

fig = px.bar(losses, x='Seleo', y='Derrotas', color_discrete_sequence=px.colors.qualitative.Bold, template='seaborn')
fig.update_layout(title = 'Selees Com Mais Derrotas')
fig.show()
```

# <font color=9453d9>5. SELEES COM MAIS GOLS MARCADOS
---

```python
goals = pd.DataFrame({'Seleo': dataset['Team'],
                      'Gols Marcados': dataset['Goals For']})
goals.sort_values('Gols Marcados', ascending=False, inplace=True)

fig = px.bar(goals, x='Seleo', y='Gols Marcados', color_discrete_sequence=px.colors.qualitative.Bold, template='seaborn')
fig.update_layout(title = 'Selees Com Mais Gols Marcados')
fig.show()
```

# <font color=9453d9>6. SELEES COM MAIS GOLS SOFRIDOS
---

```python
against = pd.DataFrame({'Seleo': dataset['Team'],
                        'Gols Sofridos': dataset['Goals Against']})
against.sort_values('Gols Sofridos', ascending=False, inplace=True)
fig = px.bar(against, 'Seleo', 'Gols Sofridos', color_discrete_sequence=px.colors.qualitative.Bold, template='seaborn')
fig.update_layout(title = 'Selees Com Mais Gols Sofridos')
fig.show()
```

# <font color=9453d9>7. SELEES COM MELHORES ATAQUES
---

```python
teams = []
games = []
goals = []

for tm in dataset['Team']:
  teams.append(tm)

for gm in dataset['Games Played']:
  games.append(gm)

for gl in dataset['Goals For']:
  goals.append(gl)

gms = np.array(games, dtype=int)
gls = np.array(goals, dtype=int)

media = gls / gms

mediaGoals = dict(zip(teams, media.tolist()))

df = pd.DataFrame(list(mediaGoals.items()),
                  columns = ['Team', 'Media Goals'])

df.sort_values('Media Goals', ascending=False, inplace=True)

fig = px.bar(df, x='Team', y='Media Goals', labels={
    'Team': 'Seleo',
    'Media Goals': 'Gols Marcados/Partidas'
}, color_discrete_sequence=px.colors.qualitative.Bold, template='seaborn')
fig.update_layout(title = 'Selees Com Melhores Ataques')
fig.show()
```

# <font color=9453d9>8. SELEES COM MELHORES DEFESAS
---

```python
teams = []
games = []
goalsAgainst = []

for tm in dataset['Team']:
  teams.append(tm)

for gm in dataset['Games Played']:
  games.append(gm)

for ga in dataset['Goals Against']:
  goalsAgainst.append(ga)

gms = np.array(games, dtype=int)
gag = np.array(goalsAgainst, dtype=int)

media = gag / gms

mediaGoalsAgainst = dict(zip(teams, media.tolist()))

df = pd.DataFrame(list(mediaGoalsAgainst.items()),
                  columns = ['Team', 'Media Goals'])

df.sort_values('Media Goals', ascending=True, inplace=True)

fig = px.bar(df, x='Team', y='Media Goals', labels={
    'Team': 'Seleo',
    'Media Goals': 'Gols Sofridos/Partidas'
}, color_discrete_sequence=px.colors.qualitative.Bold, template='seaborn')
fig.update_layout(title = 'Selees Com Melhores Defesas')
fig.show()
```

# <font color=9453d9>9. DESEMPENHO GERAL DAS SELEES
---

```python
teams = []
games = []
goals = []
goalsAgainst = []

for tm in dataset['Team']:
  teams.append(tm)

for gm in dataset['Games Played']:
  games.append(gm)

for gl in dataset['Goals For']:
  goals.append(gl)

for ga in dataset['Goals Against']:
  goalsAgainst.append(ga)

gms = np.array(games, dtype=int)
gls = np.array(goals, dtype=int)
gag = np.array(goalsAgainst, dtype=int)

media = (gls / gms) - (gag / gms)
media

mediaGoals = dict(zip(teams, media.tolist()))

df = pd.DataFrame(list(mediaGoals.items()),
                  columns = ['Team', 'Media Performance'])

df.sort_values('Media Performance', ascending=False, inplace=True)

fig = px.bar(df, x='Team', y='Media Performance', labels={
    'Team': 'Seleo',
    'Media Performance': 'Mdia De Desempenho'
}, color_discrete_sequence=px.colors.qualitative.Bold, template='seaborn')
fig.update_layout(title = 'Desempenho Geral (Desempenho=Mdia Gols Marcados-Mdia Gols Sofridos)')
fig.show()
```
